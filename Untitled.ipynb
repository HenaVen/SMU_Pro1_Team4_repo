```python
# Import Dependencies
from matplotlib import pyplot as plt
from scipy import stats
import numpy as np
import pandas as pd
```

```python
# Import the CSV
obesity_data = pd.read_csv("Data/Nutrition__Physical_Activity__and_Obesity__Behavioral_Risk_Factor_Surveillance_System.csv")
obesity_data.head()
```

```python
#Find Column names
obesity_data.columns
```

```python
# Create DF with only YearStart, LocationDesc, Topic, Data_Value_Unit, Income
obesity_df = obesity_data[["YearStart","LocationDesc", "Topic", "Data_Value", "Income"]]
obesity_df.describe()
```

```python
obesity_df_clean = obesity_df.dropna(how='any')
obesity_df_clean.head()
```

```python
obesity_df_clean.info()
```

```python
obesity_df_clean_final = obesity_df_clean[obesity_df_clean["Topic"]=="Obesity / Weight Status"]
```

```python
obesity_df_clean_final.info()
```

```python
obesity_df_clean_final.groupby(['LocationDesc'])["Data_Value"].mean()
```

```python
obesity_df_clean_final.groupby('day').filter(
    lambda x : x['total_bill'].mean() > 20
)['size'].mean()
```
